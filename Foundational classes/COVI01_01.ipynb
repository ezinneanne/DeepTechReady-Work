```python
# loading an image
import cv2
from google.colab.patches import cv_imshow
```

```python
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
DATA_DIR = '/content/drive/MyDrive/Corrupted Student Picture.png'
```

```python
# load image
img = cv2.imread(DATA_DIR)
```

```python
# image dimension
img.shape # (height, width, channel)
```

```python
plt.imshow(img) # View image with matplotlib
```

```python
cv_imshow(img) # View image with cv_imshow
```

```python
# restoring the image color
restored_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv_imshow(restored_image)
```

```python
alpha = 1.05 # Contrast control
beta = 9 # Brightness control

# call convertScaleAbs function
adjusted = cv2.convertScaleAbs(restored_image, alpha=alpha, beta=beta) # restored_image, contrast and brightness
cv_imshow(adjusted)
```

Contrast 1.05 and a brightness of 9 restored the image to its original style.
